# COVID19 Dashboard

This is a dashboard of time series data from the Johns Hopkins COVID-19 public dataset, which is updated daily with confirmed cases, deaths, and recovered counts from around the world.

## States in the United Sates

This is time series data for each state in the United States. Since there was no widespread testing in the United States prior to the first week of March, all of the time series plots start on `3/7/20'. **Note that not all states are reporting data right now, and the list below contains all of the known states as of 3/21/2020**

Despite super limited testing in the US, it's clear that every single state reporting is showing an exponential curve.

Within the United States, there are many regions, e.g., counties and cities in there as well. It's unclear to me whether the data for states is inclusive of city data, for example. This first dashboard is simply the data for the 50 states.

```python
import pandas as pd
import matplotlib.pyplot as plt

states = ['Washington', 'New York', 'California', 'Massachusetts', 'Georgia', 'Colorado', 'Florida', 'New Jersey', 'Oregon', 'Texas', 'Illinois', 'Pennsylvania', 'Iowa', 'Maryland', 'North Carolina', 'South Carolina', 'Tennessee',
         'Virginia', 'Arizona', 'Indiana', 'Kentucky', 'District of Columbia', 'Nevada', 'New Hampshire', 'Minnesota', 'Nebraska', 'Ohio', 'Rhode Island', 'Wisconsin', 'Connecticut', 'Hawaii', 'Oklahoma', 'Utah']
countries = ['US'] * len(states)
start_dates = ['3/7/20'] * len(states)
states_df = pd.DataFrame({'state': states, 'country': countries, 'start_date': start_dates})
generate_plots(states_df, 3)
plt.show()
```

## Washington State

```python
regions = ['King County, WA', 'Kittitas County, WA', 'Clark County, WA', 'Jefferson County, WA', 'Pierce County, WA', 'Grant County, WA', 'Snohomish County, WA', 'Skagit, WA', 'Thurston, WA', 'Island, WA', 'Whatcom, WA', 'Washington']
countries = ['US'] * len(regions)
start_dates = ['3/7/20'] * len(regions)
interest_df = pd.DataFrame({'state': regions, 'country': countries, 'start_date': start_dates})
generate_plots(interest_df, 3)
plt.show()
```

## Canada

It looks like BC was testing from a pretty early point in time. Most other provinces didn't start testing until much, much later. Ontario and Quebec both have rather steep slopes right now, albeit with relatively small numbers overall. But again, likely due to lack of testing.

```python
provinces = ['British Columbia', 'Alberta', 'Saskatchewan', 'Manitoba', 'Ontario', 'Quebec', 'New Brunswick', 'Nova Scotia', 'Prince Edward Island', 'Newfoundland and Labrador']
countries = ['Canada'] * len(provinces)
start_dates = ['2/1/20'] * len(provinces)
canada_df = pd.DataFrame({'state': provinces, 'country': countries, 'start_date': start_dates})
generate_plots(canada_df, 3)
plt.show()
```
