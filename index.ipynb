# COVID19 Dashboard

This is a dashboard of time series data from the [Johns Hopkins COVID-19 public dataset](https://github.com/CSSEGISandData/COVID-19), which is updated daily with confirmed cases, deaths, and recovered counts from around the world.

## North America

Exponential everywhere, and not enough testing.

```python
countries = ['US', 'Canada', 'Mexico']
states = [''] * len(countries)
start_dates = ['3/7/20'] * len(countries)
states_df = pd.DataFrame({'state': states, 'country': countries, 'start_date': start_dates})
generate_plots(states_df, 3)
plt.show()
```

## States in the United Sates

This is time series data for each state in the United States. Since there was no widespread testing in the United States prior to the first week of March, all of the time series plots start on `3/7/20'. 

Despite limited testing in the US, it's clear that every single state reporting is showing an exponential curve.

Within the United States, there are many regions, e.g., counties and cities (and cruise ships!) in there as well. The state data should be considered authoritative for cases within a state. See the appendix (likely will be moved to another notebook in the future) for an example of how this is the case for Washington State and its county data.

```python
states_df = pd.read_csv("https://raw.githubusercontent.com/jflam/covid19/master/data/states.csv")
countries = ['US'] * len(states_df)
start_dates = ['3/7/20'] * len(states_df)
states_df = pd.DataFrame({'state': states_df["State"], 'country': countries, 'start_date': start_dates})
generate_plots(states_df, 3)
plt.show()
```

## Provinces in Canada

It looks like BC was testing from a pretty early point in time. Most other provinces didn't start testing until much, much later. Ontario and Quebec both have rather steep slopes right now, albeit with relatively small numbers overall. But again, likely due to lack of testing.

```python
provinces = ['British Columbia', 'Alberta', 'Saskatchewan', 'Manitoba', 'Ontario', 'Quebec', 'New Brunswick', 'Nova Scotia', 'Prince Edward Island', 'Newfoundland and Labrador']
countries = ['Canada'] * len(provinces)
start_dates = ['2/1/20'] * len(provinces)
canada_df = pd.DataFrame({'state': provinces, 'country': countries, 'start_date': start_dates})
generate_plots(canada_df, 3)
plt.show()
```

## Selected Asian countries

It looks like China has had things under control for a while. South Korea is making good progress in turning the confirmed cases curve. Singapore and Taiwan had great early success, but it looks like visitors are driving infection rates rapidly now. 

```python
countries = ['China', 'Korea, South', 'Singapore', 'Japan', 'Taiwan*', 'India']
states = [''] * len(countries)
start_dates = ['1/22/20'] * len(countries)
asian_countries_df = pd.DataFrame({'state': states, 'country': countries, 'start_date': start_dates})
generate_plots(asian_countries_df, 3)
plt.show()
```

## Europe

The curve is exponential pretty much everywhere right now.

```python
europe_df = pd.read_csv("https://raw.githubusercontent.com/jflam/covid19/master/data/Countries-Europe.csv")
states = [''] * len(europe_df)
start_dates = ['2/21/20'] * len(europe_df)
europe_countries_df = pd.DataFrame({'state': states, 'country': europe_df['name'].sort_values(), 'start_date': start_dates})
generate_plots(europe_countries_df, 3)
plt.show()
```
